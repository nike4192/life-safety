```python
import re
from bs4 import BeautifulSoup

doc_file = open('doc.html')
html = BeautifulSoup(doc_file.read())

paragraphs = html.find_all('p')
```

```python

def oneLineQuestion(question_pieces):
    
    lines = []
    
    for question_piece in question_pieces:
        
        line = question_piece.text
        
        span_answers = question_piece.find_all('span', attrs={"class": "c13"})
        for span_answer in span_answers:
            answer = span_answer.text
            line = line.replace(answer, f'[[{answer}]]')
            
        lines.append(line)
        
    return lines


def multiLineQuestion(question_pieces):
    lines = []
    
    question = question_pieces[0].text.strip()
    lines.append(question)
    
    for i in range(1, len(question_pieces)):
        
        question_piece = question_pieces[i]
        span_answers = question_piece.find_all('span', attrs={"class": "c13"})
        
        line = str

        if question_piece.name == 'p':
            if span_answers and len(span_answers):
                line = f'+{question_piece.text.strip()}'
            else:
                line = f'-{question_piece.text.strip()}'
        else:
            line = question_piece.text
                
        lines.append(line)
                
    return lines
             
def ordererdQuestion(question_pieces):
    
    return question_pieces

def parseQuestion(question_pieces):
    
    if len(question_pieces) == 0:
        return None
    elif len(question_pieces) == 1:
        return oneLineQuestion(question_pieces)
    else:
        question = question_pieces[0]
        if question.select('.c13'):  # In question definition answer
            if not any(map(lambda p: p.select('.c13'), question_pieces[1:])):  # Has answers in list
                print('orderer questions')  # return ordererdQuestion(question_pieces)
                print(question_pieces)
            else:
                print('question_pieces')
        else:
            return multiLineQuestion(question_pieces)
```

```python
questions = []
number = -1

for el in html.find('body').contents:
    if re.match('^\s*\d+\.', el.text):
        number += 1;
        
    if number not in questions:
        questions.append(list());
        
    if re.match('^V\d+', el.text):  # V033 (Подтема): Меры по уменьшению загрязнения ОС c033 кластер (Защита селитебной территории)
        continue
        
    if el.text.strip() == '':  # empty strings
        continue
    
    questions[number].append(el)
```

```python
l = list(map(parseQuestion, questions))
```
